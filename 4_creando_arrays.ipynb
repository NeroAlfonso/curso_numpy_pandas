```python
import numpy as np
```

```python
list(range(0,10)) #crea una listaa de python en el rango indicado, excluye el último
```

```python
np.arange(0,10) #crea un arreglo de numpy con el rango suministrado
```

```python
np.arange(0,10,2) #numeros de dos en dos en el rango suministrado (step)
```

```python
np.zeros(3) #crea un arreglo de ceros con los elementos y dimensiones suministrados
```

```python
np.zeros((10,10)) #dos dimensiones de 10x10 (por ejemplo una imágen en esta resolución)
```

```python
np.ones((10,10)) #unos
```

```python
np.linspace(0,10,100) #rango inicial, rango final, número de datos, crea un arreglo distribuido entre el rango
```

```python
np.eye(10) #genera una matriz con uno en la diagonal y cero en el resto de espacios
```

```python
np.random.rand() #valor pseudoaletorio
```

```python
np.random.rand(4) #vector de 4 columnas, con valores pseudoaleatorios
```

```python
np.random.randint(1, 15) #valor aleatorio entre el rango de enteros
```
