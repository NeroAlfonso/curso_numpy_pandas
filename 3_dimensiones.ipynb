Agregar o eliminar dimensiones

```python
vector10 = np.array([1,2,3], ndmin=10) #dimensiones del array
vector10.ndim
```

```python
vector10
```

```python
vector =np.array([1,2,3])
vector.ndim
```

```python
vector_exp = np.expand_dims(vector, axis=0) #axis 0 = una dimension a nivel de filas, axis =1 columnas
vector_exp.ndim
```

```python
vector10, vector10.ndim
```

```python
vector_simp =np.squeeze(vector10) #elimina las dimensiones inutilizadas
vector_simp, vector_simp.ndim
```
