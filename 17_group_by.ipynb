```python
import pandas as pd
```

```python
df_books =pd.read_csv('https://raw.githubusercontent.com/sanketchavan5595/EDA-on-bestselling-books-on-Amazon-/main/bestsellers%20with%20categories.csv')
df_books.head
```

```python
df_books.groupby('Author').count()
```

```python
df_books.groupby('Author').sum()
```

```python
df_books.groupby('Author').max()
```

```python
df_books.groupby('Author').min()
```

```python
df_books.groupby('Author').sum().loc['Walter Isaacson']
```

```python
df_books.groupby('Author').sum().reset_index() #incorpora el indice de los elementos y no es la clave del grupo
```

```python
df_books.groupby('Author').mean().loc['Walter Isaacson'] #media y todas las funcionees de agregación pueden ser empleadas
```

```python
df_books.groupby('Author').agg(['min', 'max']) #incluye las fuciones de agregación especificadas a la vez
```

```python
df_books.groupby(['Author', 'Year']).sum() #multiples propiedades de agrupación
```

```python
df_books.groupby('Author').agg({'Reviews':'sum','User Rating':['min','max']}) #funciones de agregación especificaas por propiedad
```
