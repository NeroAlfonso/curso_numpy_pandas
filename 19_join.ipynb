```python
import pandas as pd
```

```python
df_left =pd.DataFrame({'A':['A0','A1','A2'],
  'B':['B0','B1','B2']},
  index=['k0','k1','k2'])
df_right =pd.DataFrame({
    'C':['C0','C1','C2'],
    'D':['D0','D1','D2']
}, index=['k0','k2','k3'])
```

## Join - index match

```python
df_left.join(df_right) #fusiona en funci√≥n del index
```

```python
df_left.join(df_right, how='inner')
```

```python
df_left.join(df_right, how='left')
```

```python
df_left.join(df_right, how='right')
```

```python
df_left.join(df_right, how='outer')
```
