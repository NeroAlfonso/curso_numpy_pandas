```python
import pandas as pd
```

```python
df1 =pd.DataFrame(
    {
        'A':['A0','A1','A2','A3'],
        'B':['B0','B1','B2','B3'],
        'C':['C0','C1','C2','C3'],
        'D':['D0','D1','D2','D3']
    }
)
df2 =pd.DataFrame(
    {
        'A':['A4','A5','A6','A7'],
        'B':['B4','B5','B6','B7'],
        'C':['C4','C5','C6','C7'],
        'D':['D4','D5','D6','D7']
    }
)
```

## Concat

```python
pd.concat([df1,df2]) #concatena repitiendo los indices
```

```python
pd.concat([df1,df2], ignore_index=True) #ignora el indice para crear uno sin repeticiones
```

```python
pd.concat([df1,df2], axis=1) #incluye los datos en en eje de las columnas
```

## Merge

```python
df_left =pd.DataFrame({
    'key':['k0','k1','k2','k3'],
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3']
})
df_right =pd.DataFrame({
    'key':['k0','k1','k2','k3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
})
```

```python
df_left.merge(df_right) #fusiona de forma estandar ambos DataFrames
```

```python
df_left.merge(df_right, on='key') #fusiona relacionando por el atributo key (mismo resultado)
```

```python
df_right =pd.DataFrame({
    'key_2':['k0','k1','k2','k3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
})
```

```python
#df_left.merge(df_right, on='key') #error, no existe 'key' en ambos DataFrames
```

```python
df_left.merge(df_right, left_on='key', right_on='key_2') #establece la relación explicita entre los dos DataFrames
```

```python
import numpy as np
```

```python
df_right =pd.DataFrame({
    'key_2':['k0','k1','k2',np.nan],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
})
```

```python
df_left.merge(df_right, left_on='key', right_on='key_2') #por defecto fusiona con innerjoin
```

```python
df_left.merge(df_right, left_on='key', right_on='key_2', how='left') #leftjoin
```

```python
df_left.merge(df_right, left_on='key', right_on='key_2', how='right') #rightjoin
```

```python
df_left.merge(df_right, left_on='key', right_on='key_2', how='outer') #outerjoin
```
