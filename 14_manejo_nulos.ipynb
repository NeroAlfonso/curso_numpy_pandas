```python
import pandas as pd
import numpy as np
```

```python
dic ={'col1':[1,np.nan,3,4],'col2':[10,20,30,np.nan],'col3':[None,'b','c','d']}
df =pd.DataFrame(dic)
df
```

```python
df.isnull() #indica los valores nulos
```

```python
df.isnull() *1 #sinónimo pero con numéricos
```

```python
df.fillna('Missing') #llena los valores nulos
```

```python
df.fillna(df.mean()) #llena los valores nulos con la media (promedio)
```

```python
df.interpolate() #deduce que valores deben continuar en las series
```

```python
df.dropna() #elimina los valores (y sus filas) nulos
```
